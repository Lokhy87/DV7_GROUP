```python
def clean_valor_mercado(valor):
    if isinstance(valor, str):  # Verificar si el valor es un string
        valor = valor.replace("€", "").strip()
        if "mill." in valor:
            return float(valor.replace("mill.", "").replace(",", ".")) * 1_000_000
        elif "mil" in valor:
            return float(valor.replace("mil", "").replace(",", ".")) * 1_000
    return None  # Devolver None si el formato no es esperado

# Aplicar la función de limpieza y reemplazar NaN con 0 antes de convertir a enteros
df_ligas_limpio['Valor Mercado'] = df_ligas_limpio['Valor Mercado'].apply(clean_valor_mercado).fillna(0).astype(int)

# Convertir "Edad" a numérico, reemplazar NaN con 0, y luego a entero
df_ligas_limpio['Edad'] = pd.to_numeric(df_ligas_limpio['Edad'], errors='coerce').fillna(0).astype(int)
```
