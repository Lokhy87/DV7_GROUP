HYPERMOTION

```python
df_segunda = pd.read_csv('Jugadores_LigaHypermotion.csv')
df_segunda
```

```python
df_segunda.columns = df_primera.columns.str.normalize('NFD').str.encode('ascii', 'ignore').str.decode('utf-8')
df_segunda
```

```python
df_concatenado = pd.concat([df_primera, df_segunda], axis=0)
df_concatenado
```
